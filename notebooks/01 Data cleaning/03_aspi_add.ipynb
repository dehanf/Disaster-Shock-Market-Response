```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../output/processed_data/02_suffix_handled.csv')
aspi_data = pd.read_csv('../../data/aspi.csv')
aspi_data = aspi_data.drop('volume',axis = 1)
```

```python
#Column renaming in aspi dataframe
aspi_data = aspi_data.rename(columns = {
    'open' : 'aspi_open'	,
    'high'	:'aspi_high',
    'low'	:'aspi_low',
    'close': 'aspi_close'
    
})
```

```python
print(type(df['date'][0]))
print(type(aspi_data['date'][0]))
```

```python
df = df.merge(aspi_data,on = 'date' ,how = 'left')
df.to_csv('../../output/processed_data/03_aspi_handled.csv',index = False)
```
