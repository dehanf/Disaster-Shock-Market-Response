Splitting symbol and suffix

```python
symbols = np.array(df['symbol'].str.split('.').tolist())[:,0]
suffixes = np.array(df['symbol'].str.split('.').tolist())[:,1]

df['symbol'] = symbols
df['suffix'] = suffixes
```

```python
df
```

```python
df.isnull().sum()
```

```python
sector_map = pd.read_csv('../../data/sector_mapping.csv').set_index('symbol')['sector']
df['sector'] = df['symbol'].map(sector_map).fillna('Others')

others = df[df['sector'] == 'Others']['symbol'].str.split('.').str[0].unique()
print(f"{len(others)} symbols not in mapping:\n", sorted(others))
```
