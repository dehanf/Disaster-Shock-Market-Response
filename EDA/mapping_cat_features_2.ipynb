```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../output/sector-data-imputed.csv')
```

# Mapping suffix

```python
def map_suffix(suf):
    if suf == 'N0000':  # Ordinary voting shares
        return 0
    if suf == 'X0000':  # Non-voting shares
        return 1
    if suf == 'P0000':  # Preference shares
        return 2
    if suf == 'U0000':  # Units (investment trust / fund)
        return 3
    return -1           # Unknown / other

df['suffix_cat'] = df['suffix']
df['suffix_cat'] = df['suffix_cat'].map(map_suffix)
```
