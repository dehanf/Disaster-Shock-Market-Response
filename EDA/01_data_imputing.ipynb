```python
import pandas as pd
import numpy as np
```

```python
sector_df = pd.read_csv('../data/industry_mappings.csv')
data = pd.read_csv('../data/data_with_open_prices.csv')
```

```python
sector_df
```

```python
symbols = np.array(sector_df['symbol'].str.split('.').tolist())[:,0]
suffixes = np.array(sector_df['symbol'].str.split('.').tolist())[:,1]

sector_df['symbol'] = symbols
sector_df['suffix'] = suffixes
```

```python
sector_df
```

```python
unique_symbols = sector_df['symbol'].unique().tolist()
sector_map = { symbol : sector_df[sector_df['symbol'] == symbol]['sector'].values[0]  for symbol in symbols}

```

```python
symbols = np.array(data['symbol'].str.split('.').tolist())[:,0]
suffixes = np.array(data['symbol'].str.split('.').tolist())[:,1]

data['symbol'] = symbols
data['suffix'] = suffixes
```

```python
data['sector'] = data['symbol']
data['sector'] = data['sector'].map(sector_map)
data['symbol'].nunique()
```

```python
data.to_csv('../output/01_sector-data-imputed.csv',index = False)
```
