```python
#splitting symbol into two features
symbols = np.array(data['symbol'].str.split('.').tolist())[:,0]
suffixes = np.array(data['symbol'].str.split('.').tolist())[:,1]

data['symbol'] = symbols
data['suffix'] = suffixes
```

```python
#Create a map
unique_symbols = sector_df['symbol'].unique().tolist()
sector_map = { symbol : sector_df[sector_df['symbol'] == symbol]['sector'].values[0]  for symbol in symbols}
```

```python
data.head()
```

```python
#mapping sector by symbol
data['sector'] = data['symbol']
data['sector'] = data['sector'].map(sector_map)
data['symbol'].nunique()
```
